```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC

trn = pd.read_csv("/kaggle/input/mortys-time-paradox-aicc-round-3/train.csv")
tst = pd.read_csv("/kaggle/input/mortys-time-paradox-aicc-round-3/test.csv")

X = trn.iloc[:, 1:9].to_numpy()
candidates = trn.iloc[:, 9:14].to_numpy()
X_test = tst.iloc[:, 1:9].to_numpy()
```

```python
np.random.seed(2026)

cand = candidates.astype(bool)
k = cand.sum(axis=1)

valid = k > 0
X = X[valid]
cand = cand[valid]
k = k[valid]

r = np.random.randint(0, k)
y = (cand.cumsum(axis=1) > r[:, None]).argmax(axis=1)
```

```python
model = SVC(kernel="rbf", random_state=2026)
model.fit(X, y)

y_pred = model.predict(X_test)
```

```python
pd.DataFrame({
    'row_id': tst['row_id'],
    'label': y_pred
}).to_csv('submission.csv', index=False)

print("Baseline submission.csv created.")
```
